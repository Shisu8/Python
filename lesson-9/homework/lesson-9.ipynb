5. Binary Search Tree Class
Write a Python program to create a class representing a binary search tree. Include methods for inserting and searching for elements in the binary tree.
class BST

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, key):
        if self.root is None:
            self.root = Node(key)
        else:
            self._insert_recursive(self.root, key)

    def _insert_recursive(self, current_node, key):
        if key < current_node.key:
            if current_node.left is None:
                current_node.left = Node(key)
            else:
                self._insert_recursive(current_node.left, key)
        elif key > current_node.key:
            if current_node.right is None:
                current_node.right = Node(key)
            else:
                self._insert_recursive(current_node.right, key)
        # If key is equal, it's typically ignored or handled based on specific requirements (e.g., allow duplicates)

    def search(self, key):
        return self._search_recursive(self.root, key)

    def _search_recursive(self, current_node, key):
        if current_node is None or current_node.key == key:
            return current_node
        if key < current_node.key:
            return self._search_recursive(current_node.left, key)
        else:
            return self._search_recursive(current_node.right, key)

    def inorder_traversal(self):
        result = []
        self._inorder_recursive(self.root, result)
        return result

    def _inorder_recursive(self, current_node, result):
        if current_node:
            self._inorder_recursive(current_node.left, result)
            result.append(current_node.key)
            self._inorder_recursive(current_node.right, result)
```

6. Stack Data Structure
Write a Python program to create a class representing a stack data structure. Include methods for pushing and popping elements.

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        """Adds an item to the top of the stack."""
        self.items.append(item)

    def pop(self):
        """Removes and returns the top item from the stack."""
        if not self.is_empty():
            return self.items.pop()
        else:
            return "Stack is empty"

    def peek(self):
        """Returns the top item without removing it."""
        if not self.is_empty():
            return self.items[-1]
        else:
            return "Stack is empty"

    def is_empty(self):
        """Checks if the stack is empty."""
        return len(self.items) == 0

    def size(self):
        """Returns the number of items in the stack."""
        return len(self.items)

# Example Usage:
my_stack = Stack()
my_stack.push(10)
my_stack.push(20)
my_stack.push(30)

print(f"Stack: {my_stack.items}")
print(f"Pop: {my_stack.pop()}")
print(f"Peek: {my_stack.peek()}")
print(f"Is empty: {my_stack.is_empty()}")
print(f"Size: {my_stack.size()}")
```

7. Linked List Data Structure
Write a Python program to create a class representing a linked list data structure. Include methods for displaying linked list data, inserting, and deleting nodes.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None  # Reference to the next node

class LinkedList:
    def __init__(self):
        self.head = None  # Initially, the head points to None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        last_node = self.head
        while last_node.next:
            last_node = last_node.next
        last_node.next = new_node

    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def display(self):
        current = self.head
        elements = []
        while current:
            elements.append(str(current.data))
            current = current.next
        print(" -> ".join(elements))

    def remove(self, data_to_remove):
        current = self.head
        if current and current.data == data_to_remove:
            self.head = current.next
            current = None
            return

        prev = None
        while current and current.data != data_to_remove:
            prev = current
            current = current.next

        if current is None:  # Data not found
            return

        prev.next = current.next
        current = None
```

8. Shopping Cart Class
Write a Python program to create a class representing a shopping cart. Include methods for adding and removing items, and calculating the total price.

```python
class Product:
    def __init__(self, product_id, name, price):
        self.product_id = product_id
        self.name = name
        self.price = price

    def __str__(self):
        return f"{self.name} (ID: {self.product_id}) - ${self.price:.2f}"

class ShoppingCart:
    def __init__(self):
        self.items = {}  # Dictionary to store product_id: quantity

    def add_item(self, product, quantity=1):
        if not isinstance(product, Product):
            raise TypeError("Only Product objects can be added to the cart.")
        
        if product.product_id in self.items:
            self.items[product.product_id]['quantity'] += quantity
        else:
            self.items[product.product_id] = {'product': product, 'quantity': quantity}
        print(f"Added {quantity} x {product.name} to the cart.")

    def remove_item(self, product_id, quantity=None):
        if product_id not in self.items:
            print(f"Product with ID {product_id} not found in cart.")
            return

        current_quantity = self.items[product_id]['quantity']

        if quantity is None or quantity >= current_quantity:
            del self.items[product_id]
            print(f"Removed all of product ID {product_id} from the cart.")
        else:
            self.items[product_id]['quantity'] -= quantity
            print(f"Removed {quantity} of product ID {product_id} from the cart.")

    def get_total(self):
        total = 0
        for item_data in self.items.values():
            total += item_data['product'].price * item_data['quantity']
        return total

    def display_cart(self):
        if not self.items:
            print("Your shopping cart is empty.")
            return

        print("\n--- Your Shopping Cart ---")
        for item_data in self.items.values():
            product = item_data['product']
            quantity = item_data['quantity']
            print(f"{quantity} x {product.name} (${product.price:.2f} each) = ${quantity * product.price:.2f}")
        print(f"--------------------------")
        print(f"Total: ${self.get_total():.2f}")
```

9. Stack with Display
Write a Python program to create a class representing a stack data structure. Include methods for pushing, popping, and displaying elements.

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        """Adds an item to the top of the stack."""
        self.items.append(item)

    def pop(self):
        """Removes and returns the item from the top of the stack."""
        if not self.is_empty():
            return self.items.pop()
        else:
            print("Stack is empty, cannot pop.")
            return None

    def peek(self):
        """Returns the item at the top of the stack without removing it."""
        if not self.is_empty():
            return self.items[-1]
        else:
            print("Stack is empty, no top element.")
            return None

    def is_empty(self):
        """Checks if the stack is empty."""
        return len(self.items) == 0

    def size(self):
        """Returns the number of items in the stack."""
        return len(self.items)

    def display(self):
        """Displays the current elements in the stack."""
        if not self.is_empty():
            print("Current elements in the stack (top to bottom):", self.items[::-1])
        else:
            print("Stack is empty.")

# Example Usage:
my_stack = Stack()

my_stack.push(10)
my_stack.push(20)
my_stack.push(30)
my_stack.display()

print("Top element (peek):", my_stack.peek())

popped_item = my_stack.pop()
print("Popped item:", popped_item)
my_stack.display()

my_stack.push(40)
my_stack.display()

print("Stack size:", my_stack.size())
print("Is stack empty?", my_stack.is_empty())
```


10. Queue Data Structure
Write a Python program to create a class representing a queue data structure. Include methods for enqueueing and dequeueing elements.

```python
class Queue:
    def __init__(self):
        """
        Initializes an empty Queue.
        """
        self.items = []

    def enqueue(self, item):
        """
        Adds an item to the rear of the queue.
        Time Complexity: O(1)
        """
        self.items.append(item)

    def dequeue(self):
        """
        Removes and returns the item from the front of the queue.
        Raises an IndexError if the queue is empty.
        Time Complexity: O(n) because removing from the beginning of a list
        requires shifting all subsequent elements.
        """
        if self.is_empty():
            raise IndexError("Queue is empty, cannot dequeue.")
        return self.items.pop(0)

    def front(self):
        """
        Returns the item at the front of the queue without removing it.
        Raises an IndexError if the queue is empty.
        Time Complexity: O(1)
        """
        if self.is_empty():
            raise IndexError("Queue is empty, no front element.")
        return self.items[0]

    def is_empty(self):
        """
        Checks if the queue is empty.
        Time Complexity: O(1)
        """
        return len(self.items) == 0

    def size(self):
        """
        Returns the number of items in the queue.
        Time Complexity: O(1)
        """
        return len(self.items)

    def __str__(self):
        """
        Returns a string representation of the queue.
        """
        return f"Queue({self.items})"

# Example Usage:
if __name__ == "__main__":
    my_queue = Queue()
    print(f"Is queue empty? {my_queue.is_empty()}") # True

    my_queue.enqueue(10)
    my_queue.enqueue(20)
    my_queue.enqueue(30)
    print(f"Queue after enqueues: {my_queue}") # Queue([10, 20, 30])
    print(f"Size of queue: {my_queue.size()}") # 3
    print(f"Front element: {my_queue.front()}") # 10

    dequeued_item = my_queue.dequeue()
    print(f"Dequeued item: {dequeued_item}") # 10
    print(f"Queue after dequeue: {my_queue}") # Queue([20, 30])

    print(f"Is queue empty? {my_queue.is_empty()}") # False

    my_queue.dequeue()
    my_queue.dequeue()
    print(f"Queue after all dequeues: {my_queue}") # Queue([])
    print(f"Is queue empty? {my_queue.is_empty()}") # True

    try:
        my_queue.dequeue()
    except IndexError as e:
        print(f"Error: {e}") # Error: Queue is empty, cannot dequeue.
```

```python
class Bank:
    def __init__(self, name):
        self.name = name
        self.accounts = {}  # Dictionary to store accounts, keyed by account number
        self.customers = {} # Dictionary to store customers, keyed by customer ID

    def create_account(self, customer, initial_balance=0):
        # Generate a unique account number (e.g., using a counter or random number)
        account_number = self._generate_unique_account_number()
        new_account = BankAccount(account_number, customer, initial_balance)
        self.accounts[account_number] = new_account
        # Optionally, link the account to the customer
        if customer.customer_id not in self.customers:
            self.customers[customer.customer_id] = customer
        customer.add_account(new_account)
        return new_account

    def get_account(self, account_number):
        return self.accounts.get(account_number)

    def get_customer(self, customer_id):
        return self.customers.get(customer_id)

    def _generate_unique_account_number(self):
        # Placeholder for generating a unique account number
        # In a real system, this would be more robust
        return str(len(self.accounts) + 1).zfill(5) # Example: 00001, 00002 etc.

    def transfer_funds(self, from_account_number, to_account_number, amount):
        from_account = self.get_account(from_account_number)
        to_account = self.get_account(to_account_number)

        if not from_account or not to_account:
            print("Error: One or both accounts not found.")
            return False

        if from_account.withdraw(amount):
            to_account.deposit(amount)
            print(f"Transfer of ${amount} from {from_account_number} to {to_account_number} successful.")
            return True
        else:
            print("Error: Insufficient funds or invalid withdrawal from source account.")
            return False

# Example of a BankAccount class (would typically be in a separate file or defined before Bank)
class BankAccount:
    def __init__(self, account_number, customer, initial_balance=0):
        self.account_number = account_number
        self.customer = customer
        self.balance = initial_balance
        self.transactions = []

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            self.transactions.append(f"Deposited: ${amount}")
            return True
        return False

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            self.transactions.append(f"Withdrew: ${amount}")
            return True
        return False

    def get_balance(self):
        return self.balance

    def __str__(self):
        return f"Account[{self.account_number}] - Holder: {self.customer.name} - Balance: ${self.balance}"

# Example of a Customer class
class Customer:
    def __init__(self, customer_id, name):
        self.customer_id = customer_id
        self.name = name
        self.accounts = []

    def add_account(self, account):
        self.accounts.append(account)

    def __str__(self):
        return f"Customer ID: {self.customer_id}, Name: {self.name}"
```
